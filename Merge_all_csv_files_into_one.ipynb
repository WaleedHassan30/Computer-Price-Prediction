## <center> Import Libraries

```python
# import libraries
import pandas as pd
import numpy as np
import warnings
from os import listdir
from os.path import isfile, join
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.expand_frame_repr', False)
warnings.filterwarnings("ignore")
```

<hr>

## <center> Reading All CSV File

```python
# Reading All CSV File into Variables and Create Dataframe as a Read CSV File.
path = "../../Data Collecting/Web Scraping/Downloaded_Products"
CSV_Files = sorted([f for f in listdir(path) if isfile(join(path, f))])
Var_Names = []
for i in CSV_Files:
    var = i.replace("(", "").replace(")", "").replace(".csv","").replace(" ", "_")
    locals()[var] = pd.read_csv(path+"/"+i)
    Var_Names.append(locals()[var])
    print(f"Variable {var} is assinment")
##############################################################################
```

<hr>

## <center> Merge All CSV To One CSV File

```python
# Merge All CSV To One CSV File.
# Append All Dataframes To Empaty Dataframe in Above
All_Category = pd.concat(Var_Names, ignore_index=True, axis=0)
All_Category.drop(All_Category.filter(regex="Unname"),axis=1, inplace=True)  # Remove Unnamed Column
```

```python
# Check Count of Groups
All_Category["Product_Group"].value_counts()
```

<hr>

## <center> Drop Duplicated Products

```python
# Check Duplicated Products Count
print(f"Count Of Products: {All_Category.shape[0]}")
print(f'Duplicated Products: {All_Category["Product_URL"].duplicated().sum()}')
```

```python
# Drop All Duplicated Products Rows
All_Category.drop_duplicates(subset=['Product_URL'], inplace=True)
print("Dropped")
```

```python
# ReCheck Duplicated Products Count
print(f"Count Of Products: {All_Category.shape[0]}")
print(f'Duplicated Products: {All_Category["Product_URL"].duplicated().sum()}')
```

```python
All_Category.head(2)
```

```python
# Drop Columns (Page Number - Product_no.)
All_Category.drop(["Page Number", "Product_no."],axis=1 , inplace=True)
All_Category.head(2)
```

```python
# Check Count of Groups
All_Category["Product_Group"].value_counts()
```

```python
# Save Datafame
All_Category.to_csv("CSV/df_Before_Clean.csv")
print("Saved...")
```
